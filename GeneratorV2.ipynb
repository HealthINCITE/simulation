```python
# Data files with patients, medic/treatment with cost: 
# patients, medications, immunizations, encounters, conditions, imaging, careplans, allergies
# can we merge these data files to give us one dataset?
```

```python
#First create data frames for selected datasets above
df_encounters_ma = pd.DataFrame(encounters_ma)
df_immunizations_ma = pd.DataFrame(immunizations_ma)
df_medications_ma = pd.DataFrame(medications_ma)
df_patients_ma = pd.DataFrame(patients_ma)
```

```python
display (encounters_ma.head(), immunizations_ma.head(), medications_ma.head(), patients_ma.head())
```

```python
#change column name in patient dataframe
df_patients_ma2 = df_patients_ma.rename(columns = {'Id':'PATIENT'})
```

```python
# merge ma datasets
merged_ma1 = pd.merge(df_encounters_ma, df_immunizations_ma, how = 'outer', on = 'PATIENT') 
merged_ma2 = pd.merge(df_patients_ma2, df_medications_ma, how = 'outer', on = 'PATIENT')
merged_ma3 = pd.merge(merged_ma1, merged_ma2, how = 'outer', on = 'PATIENT')
```
